

**1.We have the min and max temperatures in a city In India for each months of the year. We would like to find a function to describe this and show it graphically, the dataset
given below.**

Task:

1.fitting it to the periodic function
2.plot the fit

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
Max  = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
Min  = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
```

```python
months = np.arange(12)
plt.plot(months, Max, 'ro')
plt.plot(months, Min, 'bo')
plt.xlabel('Months')
plt.ylabel('Min & Max temperature')
```

```python
from scipy import optimize
def yearly_temps(times, avg, ampl, time_offset):
    return (avg+ ampl * np.cos((times + time_offset) * 1.8 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, months,Max, [40, 20, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months,Min, [-40, 20, 0])
```

```python
days = np.linspace(0, 12, num=365)

plt.figure(figsize=(8,6))
plt.plot(months, Max, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'r-')
plt.plot(months, Min, 'bo')
plt.plot(days, yearly_temps(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature')

plt.show()
```

**2. Charts to plot:**
   
   a. Create a pie chart presenting the male/female proportion
   
   b. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
import pandas as pd
```

```python
titanic=pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv')
```

```python
titanic.head()
```

```python
#a. Create a pie chart presenting the male/female proportion
p=titanic.sex.value_counts()
```

```python
p.plot.pie(figsize = (6,6), autopct='%1.1f%%', fontsize=15 )
plt.title('Male/Female proportion')
```

```python
titanic.sex = pd.Categorical(titanic.sex)
titanic['Sex'] = titanic.sex.cat.codes
```

```python
#b. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

titanic.plot.scatter(x='fare', y='age', c='Sex', title='scatter_plot')
plt.xlabel('Fare_paid')
plt.ylabel('Age')
```
